```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0,1,2,3,4])
y = np.array([99.96,98.95,87.88,96.39,99.98])
my_xticks = ['CS','CRT','CHAID','QUEST','INTEGRATED']
plt.title("Accuracy Comparison(Training)")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.xticks(x, my_xticks)
barlist = plt.bar(x, y)
barlist[0].set_color('red')
barlist[1].set_color('blue')
barlist[2].set_color('black')
barlist[3].set_color('black')
barlist[4].set_color('green')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0,1,2,3,4])
y = np.array([99.18,99.18,98.37,99.45,98.97])
my_xticks = ['Logistic','SVM','Decision_Tree','MLP]
plt.title("Accuracy Comparison(Testing)")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.xticks(x, my_xticks)
barlist = plt.bar(x, y)
barlist[0].set_color('red')
barlist[1].set_color('blue')
barlist[2].set_color('yellow')
barlist[3].set_color('black')
barlist[4].set_color('green')
plt.show()
```
